```python
import numpy as np
import pandas as pd

def build_frequency_table(values, interval=5):
    """
    Builds a frequency table from the given values using manual bins with a specified interval.

    Args:
        values (list or array-like): The numeric data to analyze.
        interval (int, optional): The width of each bin. Default is 5.

    Returns:
        tuple:
            - freq_df (DataFrame): Frequency table with counts and bin intervals.
            - df (DataFrame): Original data with an additional column for the bin each value belongs to.
            - bin_edges (list): List of bin edges defining the intervals.
    """
    # 1) Create a DataFrame from the input values
    df = pd.DataFrame({"Value": values})
    
    # 2) Determine the minimum and maximum values in the dataset
    min_val = int(df["Value"].min())
    max_val = int(df["Value"].max())

    # 3) Generate bin edges at specified intervals (inclusive of the max value)
    bin_edges = list(range(min_val, max_val + interval, interval))

    # 4) Assign each value to a bin using the bin edges
    #    right=False: Intervals are left-closed, right-open [a, b)
    df["Value_bin"] = pd.cut(df["Value"], bins=bin_edges, right=False)

    # 5) Count the number of values in each bin to create a frequency table
    freq_series = df["Value_bin"].value_counts().rename("Count")
    freq_df = freq_series.to_frame()

    # 6) Sort the frequency table by bin intervals for better readability
    freq_df = freq_df.sort_index()

    # 7) Optionally filter out bins with zero counts (bins with no values)
    freq_df_nonzero = freq_df[freq_df["Count"] != 0]

    # 8) Set display options to show all rows (useful for large tables)
    pd.set_option("display.max_rows", None)

    # Return the frequency table, original DataFrame with bin assignments, and bin edges
    return freq_df_nonzero, df, bin_edges

def build_frequency_table_sorted(values, interval=5):
    """
    Builds a frequency table sorted by bin intervals, including cumulative 
    counts and percentages.

    Args:
        values (list or array-like): The numeric data to analyze.
        interval (int, optional): The width of each bin. Default is 5.

    Returns:
        DataFrame: A frequency table sorted by bin intervals, with additional
                   columns for cumulative counts, cumulative percentages, and
                   percentage contributions.
    """
    # 1) Generate the frequency table and bin information
    freq_table, df, bin_edges = build_frequency_table(values, interval)

    # 2) Sort the frequency table by bin intervals for better readability
    freq_df = freq_table.sort_index()

    # 3) Compute cumulative counts and percentages
    freq_df["Cumulative"] = freq_df["Count"].cumsum()
    freq_df["Cumulative %"] = freq_df["Cumulative"] / freq_df["Count"].sum() * 100

    # 4) Calculate percentage contribution of each bin to the total count
    total_sum = freq_df["Count"].sum()
    freq_df["%"] = freq_df["Count"] / total_sum * 100

    # Return the enhanced frequency table
    return freq_df

print(build_frequency_table_sorted(values))
```

```python
import subprocess
import os

def load_git_values(repo_path):
    """
    Runs a Git command pipeline in the specified repository to extract file changes
    from the commit history and formats the output as a list of counts.

    Args:
        repo_path (str): Path to the Git repository.

    Returns:
        list: A list of integers representing the extracted counts.
    """
    try:
        # Ensure the given path is a valid Git repository
        if not os.path.isdir(repo_path):
            raise ValueError(f"The path '{repo_path}' is not a valid directory.")

        # Navigate to the repository and execute the Git command
        command = (
            'git log --numstat --pretty=format:"" | '
            'awk \'{print $3}\' | sort | uniq -c | sort -nr | awk \'{print $1 ","}\''
        )
        
        result = subprocess.run(
            command,
            shell=True,
            check=True,
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            text=True,
            cwd=repo_path  # Set the working directory to the repository path
        )
        
        # Process the output into a list of integers
        values = result.stdout.strip().split(',')
        values = [int(value) for value in values if value.strip().isdigit()]
        
        return values

    except ValueError as e:
        print("Path error:", str(e))
        return []
    except subprocess.CalledProcessError as e:
        print("Error running the Git command:", e.stderr)
        return []

# Example usage
repo_path = "/Users/guilhermebueno/Documents/Obsidian Vault/Transitional Gateway - AI"
values = load_git_values(repo_path)
freq_df = build_frequency_table_sorted(values)
print(freq_df)
```

```python
import matplotlib.pyplot as plt
def plot_cumulative_edits(freq_df):
    files = [str(interval) for interval in freq_df.index]
    
    column_names = freq_df.columns
    print(column_names)
    
    cumulative_percent = freq_df["Cumulative %"]

    plt.figure(figsize=(10, 6))
    plt.plot(files, cumulative_percent, marker='o', color='green')
    plt.xticks(rotation=45, ha='right')
    plt.title("Cumulative Distribution of Edits")
    plt.xlabel("Files (Binned)")
    plt.ylabel("Cumulative Percentage of Edits")
    plt.tight_layout()
    plt.show()
def plot_top_n_files(freq_df, n=10):
    top_n = freq_df.nlargest(n, "Count")
    files = [str(interval) for interval in top_n.index]
    counts = top_n["Count"]

    plt.figure(figsize=(10, 6))
    plt.barh(files, counts, color='orange', edgecolor='black')
    plt.gca().invert_yaxis()  # Invert y-axis for better readability
    plt.title(f"Top {n} Most Edited Files")
    plt.xlabel("Number of Edits")
    plt.ylabel("Files")
    plt.tight_layout()
    plt.show()

def plot_edits_histogram(freq_df):
    counts = freq_df["Count"]
    
    plt.figure(figsize=(10, 6))
    plt.hist(counts, bins=20, color='skyblue', edgecolor='black')
    plt.title("Distribution of Edits Across Files")
    plt.xlabel("Number of Edits")
    plt.ylabel("Number of Files")
    plt.tight_layout()
    plt.show()
    
def plot_combined_edits(freq_df):
    files = [str(interval) for interval in freq_df.index]
    counts = freq_df["Count"]
    cumulative_percent = freq_df["Cumulative %"]

    fig, ax1 = plt.subplots(figsize=(10, 6))

    # Bar chart for edit counts
    ax1.bar(files, counts, color='skyblue', edgecolor='black', label='Count')
    ax1.set_xlabel("Files (Binned)")
    ax1.set_ylabel("Number of Edits", color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')
    ax1.set_xticklabels(files, rotation=45, ha='right')

    # Line chart for cumulative percentage
    ax2 = ax1.twinx()
    ax2.plot(files, cumulative_percent, marker='o', color='green', label='Cumulative %')
    ax2.set_ylabel("Cumulative Percentage", color='green')
    ax2.tick_params(axis='y', labelcolor='green')

    plt.title("Edit Frequency and Cumulative Distribution")
    fig.tight_layout()
    plt.show()
```
